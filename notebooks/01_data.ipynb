## Đọc dữ liệu

```python
%run 00_core.ipynb
```

```python
loader = DataLoader()
prep = DataPrep()

raw = loader.load_raw()
cleaned, data_quality = prep.clean(raw)

product_catalog = prep.product_catalog(cleaned)
prior_items = prep.prior_items(cleaned)
train_items = prep.train_items(cleaned)
order_time = prep.order_time(cleaned)
data_size = prep.data_size(prior_items, train_items)
```

## Lưu dữ liệu

```python
save_csv(product_catalog, PATHS.data_file("product_catalog.csv"))
save_csv(prior_items, PATHS.data_file("prior_items.csv"))
save_csv(train_items, PATHS.data_file("train_items.csv"))
save_csv(order_time, PATHS.data_file("order_time.csv"))
save_csv(data_quality, PATHS.table_file("data_quality.csv"))
save_csv(data_size, PATHS.table_file("data_size.csv"))
```
