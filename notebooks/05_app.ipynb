```python
import subprocess
import sys
import socket
from pathlib import Path
```

```python
app_path = Path("..") / "app" / "streamlit_app.py"

if not app_path.exists():
    raise FileNotFoundError(f"Không thấy file tại: {app_path.resolve()}")

print(f"App file: {app_path}")
```

## Mở app

```python
def get_free_port():
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sock:
        sock.bind(("0.0.0.0", 0))
        return sock.getsockname()[1]

free_port = get_free_port()

streamlit_process = subprocess.Popen(
    [
        sys.executable,
        "-m",
        "streamlit",
        "run",
        str(app_path),
        "--server.port",
        str(free_port),
        "--server.address",
        "0.0.0.0",
        "--server.headless",
        "true",
    ]
)

print(f"Link truy cập: http://localhost:{free_port}")
```
